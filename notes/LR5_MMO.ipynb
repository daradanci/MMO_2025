```python
text = 'Как видите, мы за Вами давненько наблюдаем, мистер Андерсон. Оказывается, Вы живёте двойной жизнью. В одной жизни Вы — Томас Андерсон, программист в одной крупной уважаемой компании. У Вас есть медицинская страховка, Вы платите налоги и ещё — помогаете консьержке выносить мусор. Другая Ваша жизнь — в компьютерах, и тут Вы известны как хакер Нео. Вы виновны практически во всех известных компьютерных преступлениях.'
```

```python
!pip install -U spacy
!python -m spacy download ru_core_news_sm
```

```python
from spacy.lang.ru import Russian
import spacy
nlp = spacy.load('ru_core_news_sm')
spacy_text1 = nlp(text)
spacy_text1
```

```python
for t in spacy_text1:
    print(t)
```

**Частеречная разметка**

```python
for token in spacy_text1:
    print('{} - {} - {}'.format(token.text, token.pos_, token.dep_))
```

# Лемматизация

```python
for token in spacy_text1:
      print(token, token.lemma, token.lemma_)
```

# Выделение (распознавание) именованных сущностей, named-entity recognition

```python
for ent in spacy_text1.ents:
    print(ent.text, ent.label_)
```

```python
from spacy import displacy
displacy.render(spacy_text1, style='ent', jupyter=True)
```
